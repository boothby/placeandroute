```python
from Embedding import *
import networkx as nx
```

```python
G = nx.Graph()
G.add_nodes_from([1,2])
G.add_edges_from([(1,2)])
A = nx.Graph()
A.add_nodes_from([1,2])
A.add_edge(1,2)
G, _ = create2(2,2)
A, l = create(8,8)
embed(G,A)
```

```python
from bokeh.plotting import output_notebook, figure, show
output_notebook()
```

```python
f = figure()
nodesx, nodesy = [], []

for node, datum in A.nodes_iter(data=True):
    #f.circle(x=datum["x"], y=datum["y"], radius=0.1)
    nodesx.append(datum["x"])
    nodesy.append(datum["y"])
    if "mapped" in datum and len(datum["mapped"]) >= 1:
        f.circle(datum["x"], datum["y"], radius=0.3)

f.circle(nodesx, nodesy, radius = 0.1)

edgesposx = []
edgesposy = []
for node1, node2 in A.edges_iter():
    datum1 = A.node[node1]
    datum2 = A.node[node2]
    #f.line([datum1["x"], datum2["x"]], [datum1["y"], datum2["y"]], line_width=0.1)
    edgesposx.append([datum1["x"], datum2["x"]])
    edgesposy.append([datum1["y"], datum2["y"]])
f.multi_line(edgesposx, edgesposy, line_width=0.2)
f.xgrid.grid_line_color = None
f.ygrid.grid_line_color = None
f.axis.axis_label_text_color = None
f.axis.axis_line_color = None
f.axis.major_label_text_color = None
f.axis.major_tick_line_color = None
f.axis.minor_tick_line_color = None

show(f)
```
